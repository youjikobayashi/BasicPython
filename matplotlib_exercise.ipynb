## 1問目
使うcsvはhist_data.csvです。下にコードを書いてください。

```python
temperature = pd.read_csv("data/hist_data.csv")


        
plt.title("aデータとbデータの温度に関する度数分布", fontsize=12) 
plt.ylabel("度数", fontsize=10 ) 
plt.xlabel("温度(℃)", fontsize=10)
plt.grid(False)                           
plt.hist(temperature["aデータ"], bins=30, alpha=0.6, label=["aデータ"], color="#FFD17F")
plt.hist(temperature["bデータ"], bins=30, alpha=0.6, label=["bデータ"], color="#9FEFE7")
plt.legend(loc="upper right")
plt.show()
```

## 2問目
使うcsvはscatter_data.csvです。下にコードを書いてください。

![問2](./imgs/matplotlib/plot_2.png)

```python
length_width_scatter = pd.read_csv("data/scatter_data.csv")

plt.title("クラス0～2のlengthとwidthに関する散布図")
plt.xlabel("length (cm)")
plt.ylabel("width (cm)")
plt.grid(True)
length_width_scatter_class0 = length_width_scatter[length_width_scatter["class"] == 0]
length_width_scatter_class1 = length_width_scatter[length_width_scatter["class"] == 1]
length_width_scatter_class2 = length_width_scatter[length_width_scatter["class"] == 2]
plt.scatter(length_width_scatter_class0["length (cm)"], length_width_scatter_class0["width (cm)"], c="Red", label="class0")
plt.scatter(length_width_scatter_class1["length (cm)"], length_width_scatter_class1["width (cm)"], c="Blue", label="class1")
plt.scatter(length_width_scatter_class2["length (cm)"], length_width_scatter_class2["width (cm)"], c="Green", label="class2")
plt.legend(loc="upper right")
```

## 3問目
使うcsvはline_data.csvです。下に2つの画像を表示するコードを書いてください。

![問3_1](./imgs/matplotlib/plot_3_1.png)

![問3_2](./imgs/matplotlib/plot_3_2.png)

```python
number_periods_costs = pd.read_csv("data/line_data.csv")
fig = plt.figure(
    figsize=(6, 9)
)
ax1 = fig.add_subplot(
    211,
    title="期間ごとの件数の推移",
    xlabel="期間",
    ylabel="件数",
    )
ax2 = fig.add_subplot(
    212,
    title="期間ごとの費用の推移",
    xlabel="期間",
    ylabel="費用(円)",
    )
ax1.grid(True)
ax1.set_xticklabels(number_periods_costs["期間"], rotation=30, ha="center")
ax1.plot(number_periods_costs["期間"],number_periods_costs["件数"], marker="o")

ax2.set_xticklabels(number_periods_costs["期間"], rotation=30, ha="center")
ax2.grid(True)
ax2.bar(number_periods_costs["期間"], number_periods_costs["費用"])

fig.tight_layout()
```
